```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Population.csv")
```

```python
df.drop([14, 15, 16], inplace=True)
```

```python
# Draw line & bar charts to show urban and rural population of India over the years using population.csv 
# [rural population is the difference between total population and urban population]
# [For bar chart consider 8 yrs data]


df['Population'] = pd.to_numeric(df['Population'], errors='coerce')
df['Urban Population'] = pd.to_numeric(df['Urban Population'], errors='coerce')

df['Rural Population'] = df['Population'] - df['Urban Population']

#LINE CHART
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['Urban Population'], label='Urban Population', marker='o')
plt.plot(df['Year'], df['Rural Population'], label='Rural Population', marker='o')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Urban and Rural Population of India Over the Years')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#BAR CHARtT 8 Years
df_subset = df.tail(8)

plt.figure(figsize=(12, 6))
X = df_subset['Year']
plt.bar(X - 0.2, df_subset['Urban Population'], width=0.4, label='Urban Population')
plt.bar(X + 0.2, df_subset['Rural Population'], width=0.4, label='Rural Population')

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Urban and Rural Population of India (Last 8 Years)')
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.savefig('urban_rural_population_bar_chart.png')
plt.show()
```

```python
# Using gapminder.csv show life expectancy of top 10 highly populated countries from Asia
gapminder = pd.read_csv("gapminder2007 (1).csv") 
asia = gapminder[gapminder['continent'] == 'Asia']

top10_pop = asia.sort_values(by='pop', ascending=False).head(10)
plt.figure(figsize=(12, 6))
plt.bar(top10_pop['country'], top10_pop['lifeExp'], color='orange')
plt.xlabel('Country')
plt.ylabel('Life Expectancy (Years)')
plt.title('Life Expectancy of Top 10 Most Populated Asian Countries (2007)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Create sample data for top 6 selling cars
car_data = {
    'Car Model': ['Maruti Swift', 'Hyundai i20', 'Tata Nexon', 'Kia Seltos', 'Mahindra Scorpio', 'Toyota Innova'],
    'Sales': [25000, 18000, 22000, 17000, 15000, 20000]
}

car_df = pd.DataFrame(car_data)

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(car_df['Sales'], labels=car_df['Car Model'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Top 6 Selling Cars')
plt.tight_layout()
plt.show()
```

```python
# Using Histogram show distribution of developers by their experience [use survey_data_sample.csv]
survey_data = pd.read_csv("survey_results_sample.csv")

survey_data['YearsCode'] = (
    survey_data['YearsCode']
    .replace({'Less than 1 year': '0', 'More than 50 years': '51'})
)
survey_data['YearsCode'] = pd.to_numeric(survey_data['YearsCode'], errors='coerce')

survey_data = survey_data.dropna(subset=['YearsCode'])

#histogram
plt.figure(figsize=(10, 6))
plt.hist(survey_data['YearsCode'], bins=10, edgecolor='black')
plt.title('Distribution of Developers by Coding Experience')
plt.xlabel('Years of Coding Experience')
plt.ylabel('Number of Developers')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
